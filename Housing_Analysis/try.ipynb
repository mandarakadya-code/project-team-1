```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'VALUE' is the column containing housing price indexes
# and 'GEO' is the column containing regions

# List of regions to plot
regions_to_plot =['Canada', 'Quebec', 'Atlantic Region', 'Manitoba', 'Ontario']

# Filter the data for the specified regions
filtered_data = Hou2_df[Hou2_df['GEO'].isin(regions_to_plot)]

# Create a box plot
plt.figure(figsize=(12, 8))

# Use the boxplot function and specify x and y data
plt.boxplot([filtered_data[filtered_data['GEO'] == region]['VALUE'] for region in regions_to_plot], labels=regions_to_plot, patch_artist=True)

# Add a title and labels
plt.title('Housing Price Index Distribution by Region (Box Plot)')
plt.xlabel('Region')
plt.ylabel('Housing Price Index')

# Add grid lines for better readability
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Customize box colors
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(plt.boxplot([filtered_data[filtered_data['GEO'] == region]['VALUE'] for region in regions_to_plot], labels=regions_to_plot)['boxes'], colors):
    patch.set_facecolor(color)

# Show the plot
plt.savefig('histo.png')
plt.show()
```

```python


# List of regions to plot
regions_to_plot = ['Canada', 'Quebec', 'Atlantic Region', 'Manitoba', 'Ontario']

# Filter the data for the specified regions
filtered_data = Hou2_df[Hou2_df['GEO'].isin(regions_to_plot)].copy()  
# Convert 'REF_DATE' column to datetime format
filtered_data['REF_DATE'] = pd.to_datetime(filtered_data['REF_DATE'])

# Extract year from 'REF_DATE'
filtered_data['Year'] = filtered_data['REF_DATE'].dt.year

# Create a box plot by year
plt.figure(figsize=(12, 8))

# Use the boxplot function and specify x and y data
boxprops = dict(facecolor='lightblue', color='black')  # Box properties
plt.boxplot([filtered_data[filtered_data['GEO'] == region].groupby('Year')['VALUE'].mean() for region in regions_to_plot],
            labels=regions_to_plot, patch_artist=True, boxprops=boxprops)

# Add a title and labels
plt.title('Average Housing Price Index by Year (Box Plot)')
plt.xlabel('Region')
plt.ylabel('Average Housing Price Index')

# Add grid lines for better readability
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
